Supervised learning > Regression > basic liner regression using single variable

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Sample data: house sizes (in square feet) and their corresponding prices
house_sizes = np.array([1000, 1500, 1800, 2500, 3000, 3500, 4000]).reshape(-1, 1)  # Reshape to make it a 2D array
house_prices = np.array([300000, 450000, 500000, 600000, 700000, 800000, 900000])
```

```python
# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(house_sizes, house_prices)
```

```python
# Predict for a new house size
new_house_size = np.array([[2800]])  # Predicting for a 2800 sq. ft. house
predicted_price = model.predict(new_house_size)

print(f"Predicted price for a 2800 sq. ft. house: ${predicted_price[0]:,.2f}")

# Plotting the results
plt.scatter(house_sizes, house_prices, color='blue', label='Actual data')
plt.plot(house_sizes, model.predict(house_sizes), color='red', label='Linear regression model')
plt.scatter(new_house_size, predicted_price, color='green', marker='*', s=200, label='Predicted house')
plt.xlabel('House Size (sq. ft.)')
plt.ylabel('Price ($)')
plt.title('House Price Prediction with Linear Regression')
plt.legend()
plt.show()
```
