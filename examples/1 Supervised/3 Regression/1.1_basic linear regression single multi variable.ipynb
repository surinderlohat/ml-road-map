Supervised learning > Regression > basic liner regression using single variable

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Sample data: house sizes (in square feet) and their corresponding prices
house_sizes = np.array([1000, 1500, 1800, 2500, 3000, 3500, 4000]).reshape(-1, 1)  # Reshape to make it a 2D array
house_prices = np.array([300000, 450000, 500000, 600000, 700000, 800000, 900000])
```

```python
house_sizes
```

```python
# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(house_sizes, house_prices)
```

```python
# Predict for a new house size
new_house_size = np.array([[2800]])  # Predicting for a 2800 sq. ft. house
predicted_price = model.predict(new_house_size)

print(f"Predicted price for a 2800 sq. ft. house: ${predicted_price[0]:,.2f}")

# Plotting the results
plt.scatter(house_sizes, house_prices, color='blue', label='Actual data')
plt.plot(house_sizes, model.predict(house_sizes), color='red', label='Linear regression model')
plt.scatter(new_house_size, predicted_price, color='green', marker='*', s=200, label='Predicted house')
plt.xlabel('House Size (sq. ft.)')
plt.ylabel('Price ($)')
plt.title('House Price Prediction with Linear Regression')
plt.legend()
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data: advertising spending (in thousands of dollars) and corresponding sales (in thousands of units)
advertising_spends = np.array([
    [100, 20, 10],
    [200, 30, 20],
    [300, 40, 25],
    [400, 50, 30],
    [500, 60, 35]
])

sales = np.array([15, 25, 35, 45, 55])

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(advertising_spends, sales)

# Predict for new advertising spends
new_advertising_spends = np.array([[250, 35, 15]])  # Predicting for spends: TV=250k, Radio=35k, Newspaper=15k
predicted_sales = model.predict(new_advertising_spends)

print(f"Predicted sales for advertising spends: TV=$250k, Radio=$35k, Newspaper=$15k: {predicted_sales[0]:,.2f} units")

# Plotting the results
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Plotting TV vs. Sales
axs[0].scatter(advertising_spends[:, 0], sales, color='blue', label='Actual data')
axs[0].scatter(new_advertising_spends[0, 0], predicted_sales, color='green', marker='*', s=200, label='Predicted point')
axs[0].set_xlabel('TV Advertising Spend (k$)')
axs[0].set_ylabel('Sales (k units)')
axs[0].set_title('TV vs. Sales')

# Plotting Radio vs. Sales
axs[1].scatter(advertising_spends[:, 1], sales, color='blue', label='Actual data')
axs[1].scatter(new_advertising_spends[0, 1], predicted_sales, color='green', marker='*', s=200, label='Predicted point')
axs[1].set_xlabel('Radio Advertising Spend (k$)')
axs[1].set_ylabel('Sales (k units)')
axs[1].set_title('Radio vs. Sales')

# Plotting Newspaper vs. Sales
axs[2].scatter(advertising_spends[:, 2], sales, color='blue', label='Actual data')
axs[2].scatter(new_advertising_spends[0, 2], predicted_sales, color='green', marker='*', s=200, label='Predicted point')
axs[2].set_xlabel('Newspaper Advertising Spend (k$)')
axs[2].set_ylabel('Sales (k units)')
axs[2].set_title('Newspaper vs. Sales')

plt.tight_layout()
plt.show()
```
