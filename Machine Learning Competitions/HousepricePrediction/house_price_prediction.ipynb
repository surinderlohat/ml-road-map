```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# Load CSV Data into pandas data frame
training_df = pd.read_csv("./train.csv")
testing_df = pd.read_csv("./train.csv")
```

```python
# see how our data is looks like
training_df.head()
```

```python
# now we know machine not understand the string values we first we need to convert all our string features into numbers
def pre_process_data(df):
    temp_df = pd.DataFrame(df)
    # Encode the target variable
    label_encoder = LabelEncoder()
    # Map All class based columns into numeric format
    for column in temp_df.columns:
        # unique data after removing na
        unique_records = temp_df[column].dropna().unique()

        # if column contain string type values then we needs to convert it into numeric values
        is_string_type = pd.api.types.is_string_dtype(unique_records)
        data = temp_df[column]
        if is_string_type:
            temp_df[column] = label_encoder.fit_transform(data)
    return temp_df

```

```python
# pre process data using above method
training_new_df = pre_process_data(training_df)

# save levels
y_train = training_new_df['SalePrice']
# remove sale price column
x_train = training_new_df.drop('SalePrice', axis=1)

training_new_df.head()
```

```python
# pre process data using above method
x_testing_new_df = pre_process_data(testing_df)
# save levels
y_test = x_testing_new_df['SalePrice']
# remove sale price column
x_test = x_testing_new_df.drop('SalePrice', axis=1)
x_testing_new_df.head()
```

We will store the Y levels data into y_salePrice and after that drop the sale price column from or data frame

```python
x_test.shape[1]
```

```python
# Build the neural network model
model = tf.keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(x_test.shape[1],)),
    layers.Dropout(0.2),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(1)  # Output layer
])

# Compile the model
# model.compile(optimizer='adam', loss='mse', metrics=['mae'])
# Compile the model with a lower learning rate
model.compile(optimizer=tf.keras.optimizers.Adam(
    learning_rate=0.001), loss='mse', metrics=['mae'])
```

```python
# Evaluate the model on the test set
test_loss, test_mae = model.evaluate(x_test, y_test, verbose=2)
print(f"Test MAE: {test_mae}")
```
